```python
# tic tac toe - version 2

board = [" "," "," "," "," "," "," "," "," "," "] # board is stored in a list
player = 1 # assumes player 1 always starts

# random variables to store the game result
Win = 2
Draw = 3
Progress = 4

GameRunning = True # assumes that the game is always running
GameResult = Progress # assumes there is no result yet as the game is still running

def DrawBoard(board): # board is put in a function in order to save space
    print(board[1], '|', board[2], '|', board[3])
    print('---------')
    print(board[4], '|', board[5], '|', board[6])
    print('---------')
    print(board[7], '|', board[8], '|', board[9])
    print()

def CheckMark(x):
    if(board[x] == " "): # checks whether there is a space availiable
        return True # if so, the game will carry on running
    else:
        print("Error - this space is already taken") # if not, this error message will be printed
        print()
        return False # and the game will ask the player to choose another location

def CheckWin():
    # checks if a player has won - diagonally
    global GameResult
    if board[1] == board[2] and board[2] == board[3] and board[1] != " ":
        GameResult = Win # tells the computer that someone has won
        return False # game will stop running
    elif board[4] == board[5] and board[5] == board[6] and board[4] != " ":
        GameResult = Win
        return False
    elif board[7] == board[8] and board[8] == board[9] and board[7] != " ":
        GameResult = Win
        return False
    # checks if a player has won - vertically
    elif board[1] == board[4] and board[4] == board[7] and board[1] != " ":
        GameResult = Win
        return False
    elif board[2] == board[5] and board[5] == board[8] and board[2] != " ":
        GameResult = Win
        return False
    elif board[3] == board[6] and board[6] == board[9] and board[3] != " ":
        GameResult = Win
        return False
    # checks if a player has won - diagonally
    elif board[1] == board[5] and board[5] == board[9] and board[1] != " ":
        GameResult = Win
        return False
    elif board[3] == board[5] and board[5] == board[7] and board[3] != " ":
        GameResult = Win
        return False
    # checks if no one has won and it is a tie
    elif board[1] != " " and board[2] != " " and board[3] != " " and board[4] != " " and board[5] != " " and board[6] != " " and board[7] != " " and board[8] != " " and board[9] != " ": 
        GameResult = Draw
    else:
        return True # game will carry on running if no one wins or the game is a tie

print("Player 1 = X and Player 2 = O") # indicates to the users which character they will represent
print()
        
while GameRunning == True:
    DrawBoard(board) # prints an updated grid every time a player inputs a new position
    if player%2 != 0: # checks which player goes first
        print("Player 1's Turn...") # indicates to the users who should currently be inputting their position
        Mark = "X"
    else:
        print("Player 2's Turn...")
        Mark = "O"
    choice = int(input("Enter your position between 1-9: ")) # allows the players to input their position, using only integers
    print()
    if(CheckMark(choice)):
        board[choice] = Mark # game will input the player's chosen position onto the board
        player+=1
        GameRunning = CheckWin() # game will check if anyone has won

if GameResult == Draw:
    DrawBoard(board)
    print("It is a draw") # indicates to the users that the game has tied
elif GameResult == Win:
    if player % 2 != 0: # identifies which player has won by calculating the number of characters that are currently stored on the board
        DrawBoard(board) # displays to the users the final outcome of the game
        print("Player 2 has won.") # indicates to the users who have won
    else:
        DrawBoard(board)
        print("Player 1 has won.")
        
# possible enhancements for version 3 :
# - add a visual guide for players to help them choose their position
# - replace the random variables of game results with a point system
# - ask players if they would like to play again
```
