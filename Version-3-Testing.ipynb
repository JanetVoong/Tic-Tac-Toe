```python
# tic tac toe - version 3

board = [" "," "," "," "," "," "," "," "," "," "] # board is stored in a list
player = 1 # assumes player 1 always starts
positions = [1, 2, 3, 4, 5, 6, 7, 8, 9]

# point system
Win = +1
Draw = -1
Progress = 0

GameRunning = True # assumes that the game is always running
GameResult = Progress # assumes there is no result yet as the game is still running

def DrawBoard(board): # board is put in a function in order to save space
    print(board[1], '|', board[2], '|', board[3])
    print('---------')
    print(board[4], '|', board[5], '|', board[6])
    print('---------')
    print(board[7], '|', board[8], '|', board[9])
    print()

def GuideBoard(board): # guide of the board is put in a function in order to save space
    print("1", '|', "2", '|', "3")
    print('---------')
    print("4", '|', "5", '|', "6")
    print('---------')
    print("7", '|', "8", '|', "9")
    print()

def CheckMark(x):
    if(board[x] == " "): # checks whether there is a space availiable
        return True # if so, the game will carry on running
    else:
        print("Error - this space is already taken") # if not, this error message will be printed
        print()
        return False # and the game will ask the player to choose another location

def CheckWin():
    # checks if a player has won - horizontally
    global GameResult
    if board[1] == board[2] and board[2] == board[3] and board[1] != " ":
        GameResult = Win # tells the computer that someone has won
        return False # game will stop running
    elif board[4] == board[5] and board[5] == board[6] and board[4] != " ":
        GameResult = Win
        return False
    elif board[7] == board[8] and board[8] == board[9] and board[7] != " ":
        GameResult = Win
        return False
    # checks if a player has won - vertically
    elif board[1] == board[4] and board[4] == board[7] and board[1] != " ":
        GameResult = Win
        return False
    elif board[2] == board[5] and board[5] == board[8] and board[2] != " ":
        GameResult = Win
        return False
    elif board[3] == board[6] and board[6] == board[9] and board[3] != " ":
        GameResult = Win
        return False
    # checks if a player has won - diagonally
    elif board[1] == board[5] and board[5] == board[9] and board[1] != " ":
        GameResult = Win
        return False
    elif board[3] == board[5] and board[5] == board[7] and board[3] != " ":
        GameResult = Win
        return False
    # checks if the game is a tie
    elif board[1] != " " and board[2] != " " and board[3] != " " and board[4] != " " and board[5] != " " and board[6] != " " and board[7] != " " and board[8] != " " and board[9] != " ": 
        GameResult = Draw
    else:
        return True # game will carry on running if no one wins or the game is not a tie

def RunGame(GameRunning): # main code for the game is put in a function so that it is able to loop
    player = 1 # player 1 will always start first
    while GameRunning == True:
        DrawBoard(board) # prints an updated grid every time a player inputs a new position
        if player%2 != 0: # checks which player goes first
            print("Player 1's Turn...") # indicates to the users who should currently be inputting their position
            Mark = "X" # player 1 will always represent "X"
        else:
            print("Player 2's Turn...")
            Mark = "O" # player 2 will always represent "O"
        choice = int(input("Enter your position between 1-9: ")) # allows the players to input their position, using only integers
        #print()
        #if(CheckMark(choice)):
            #board[choice] = Mark # game will input the player's chosen position onto the board
            #player+=1
            #GameRunning = CheckWin() # game will check if anyone has won
        
        for i in positions:
            if choice != i:
                print("Error - enter a number between 1-9.")

def CheckResult(board):
    # stores players' points
    player1Points = 0 # assumes the players have no points at the start of the game
    player2Points = 0
    if GameResult == Draw:
        DrawBoard(board)
        print("It is a draw.") # indicates to the users that the game has tied
        player1Points+=Draw # the game will minus 1 point from each of the players
        player2Points+=Draw
    elif GameResult == Win:
        if player % 2 == 0: # identifies which player has won by calculating the number of characters that are currently stored on the board
            DrawBoard(board) # displays to the users the final outcome of the game
            print("Player 2 has won.") # indicates to the users who have won
            player2Points+=Win # the game will add 1 point to the winning player
        else:
            DrawBoard(board)
            print("Player 1 has won.")
            player1Points+=Win
    print("Player 1's Points:", player1Points, "| Player 2's Points:", player2Points) # indicated to the players how many points they currently have
    print()

def CheckPlayAgain(): # I put this check into a function
    playAgain = input("Would you like to play again? Y/N: ").upper() # Y/N would reduce the inaccuracy of data, also making the input always uppercase
    if playAgain == "Y":
        print()
        board = [" "," "," "," "," "," "," "," "," "," "] # resets the board
        RunGame(GameRunning = True) # loops the game
        CheckResult(board)
    elif playAgain == "N":
        print("Thanks for playing!")
    else:
        print("Something went wrong, please try again.")
        print()
        return CheckPlayAgain() # this will enable the game to loop back to the question
            
print("Welcome to Tic-Tac-Toe by Janet Voong") # shows players who created the game
print()
print("This is a guide:") # displays a guide to help players choose their position
print()
GuideBoard(board)
print("Player 1 = X and Player 2 = O") # indicates to the players which character they will represent
print()
RunGame(GameRunning)
print()
CheckResult(board)
print()
CheckPlayAgain()
print()
```
